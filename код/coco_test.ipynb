# Requirements:

```python
%matplotlib inline
from pycocotools.coco import COCO
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
import pylab
pylab.rcParams['figure.figsize'] = (8.0, 10.0)
```

# Importing COCO JSON file:

```python
coco = COCO("/content/drive/MyDrive/proj_test/jsonfile.json")
```

# Categories:

```python
cats = coco.loadCats(coco.getCatIds())
nms=[cat['name'] for cat in cats]
print('COCO categories: \n{}\n'.format(' '.join(nms)))
```

# Load and display image:

```python
catIds = coco.getCatIds(catNms=['face','left_eye','right_eye']);
imgIds = coco.getImgIds(catIds=catIds);
img = coco.loadImgs(imgIds)[0]
catIds
```

```python
I = io.imread("/content/drive/MyDrive/proj_test/1.png")
plt.axis('off')
plt.imshow(I)
plt.show()
```

# Load and display instance annotations (**without bbox**)

```python
plt.imshow(I); plt.axis('off')
annIds = coco.getAnnIds(imgIds=1, catIds=catIds, iscrowd=None)
anns = coco.loadAnns(annIds)
coco.showAnns(anns)
coco.showAnns(anns, draw_bbox = False)
```

# Load and display instance annotations (**with bbox**)

```python
plt.imshow(I); plt.axis('off')
coco.showAnns(anns, draw_bbox = True)
```
